```python
preds = rf.predict(test[predictors])
```

```python
def rolling_averages(group, columns, new_columns):
  group = group.sort_values("date")
  rolling_stats = group[columns].rolling(3, closed='left').mean()
  group[new_columns] = rolling_stats
  group = group.dropna(subset=new_columns)
  return group
```

```python
columns = ["GF", "GA", "Sh", "SoT", "Dist", "FK", "PK", "PKatt"]
new_columns = [f"{c}_rolling" for c in columns]
```

```python
new_columns
```

```python
group.columns
```

```python
rolling_averages(group, columns, new_columns)
```

```python
matches_rolling = matches.groupby("Team").apply(lambda x: rolling_averages(x, columns, new_columns))
```

```python
matches_rolling
```

```python
matches_rolling = matches_rolling.droplevel("Team")
```

```python
matches_rolling
```

```python
matches_rolling.index = range(matches_rolling.shape[0])
```

```python
matches_rolling
```

```python
def make_predictions(data, predictors):
    train = data[data["date"] < '2022-01-01']
    test = data[data["date"] >= '2022-01-01']
    rf.fit(train[predictors], train["target"])
    preds = rf.predict(test[predictors])
    combined = pd.DataFrame(dict(actual=test["target"], predicted=preds), index=test.index)
    error = precision_score(test["target"], preds)
    return combined, error
```

```python
combined, precision = make_predictions(matches_rolling, predictors + new_columns)
```
